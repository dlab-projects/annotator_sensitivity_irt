```python
import numpy as np
import pandas as pd
import seaborn as sns


import csv
import os 

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_multi = pd.read_feather('/home/rbarreto/data/scaled-facets-multi-50k-v4.feather')
data_multi.head(5)
```

```python
number_labelers = data_multi['comment_id'].value_counts()
```

```python
comment_ids_reference = number_labelers[number_labelers > 50].index
```

```python
df_reference = data_multi[data_multi['comment_id'].isin(comment_ids_reference)]
```

```python
#reconstructing hate speech sum

# Hate speech construct items
items = [
    "sentiment",
    "respect",
    "insult",
    "humiliate",
    "status",
    "dehumanize",
    "violence_phys",
    "genocide",
    "attack_defend",
    "hatespeech"]

hate_speech_sum = df_reference[items].sum(axis=1)

df_reference['hate_speech_sum'] = hate_speech_sum

df_reference
```

```python
#dont wan't samples in df that have other races bc just black and white annotators

col_to_annotator_race = {
    'demo_race_ethnicitie_1': 'native_american',
    'demo_race_ethnicitie_2': 'asian',
    'demo_race_ethnicitie_3': 'black',
    'demo_race_ethnicitie_4': 'latinx',
    'demo_race_ethnicitie_5': 'pacific_islander',
    'demo_race_ethnicitie_6': 'white',
    'demo_race_ethnicitie_7': 'other race',
    'demo_race_ethnicitie_8': 'middle eastern'}

def swap_key_val(dic):
  """Swaps key, value pairs in a dictionary."""
  return {val: key for key, val in dic.items()}

annotator_race_to_col = swap_key_val(col_to_annotator_race)

annotator_race_to_col
```

```python
#black annotators all of them that are in the reference set 
black_white_reference = df_reference[(df_reference[annotator_race_to_col['black']] == 1) | 
                                     (df_reference[annotator_race_to_col['white']] == 1)]

black_white_reference.shape
```

```python
annotator_agreement = black_white_reference[['comment_id', 'target_race_1']].fillna(0).groupby('comment_id').mean()
```

```python
#75% of annotators agree that this targets black folks
black_target_idx = annotator_agreement[annotator_agreement['target_race_1'] > 0.75].index
```

```python
black_targeting_comments = black_white_reference[black_white_reference['comment_id'].isin(black_target_idx)]


black_targeting_comments.shape

#8426 samples comprised of 16 comments
```

```python
#only 16 comments are in the reference set, have black and white annotators ,and target ppl on the basis of race so that most (75%) annotators agree that it's targeted on basis of race (black)
black_target_idx.shape
```
